```python
import cv2
import numpy as np
from skimage import io
```

```python
img = io.imread('C:/Users/uib43225/PycharmProjects/DSAlgo/computer vision/cat_224x224.jpg')
io.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
io.imshow(gray)
```

```python
img_gaussian = cv2.GaussianBlur(gray,(3,3),0)
io.imshow(img_gaussian)
```

```python
img_canny = cv2.Canny(img,100,200)
io.imshow(img_canny)
```

```python
#sobel
#It is mostly in the case of image-noise, second derivative operator.
#It gives dense edges, It is much more able to give extracted information.
img_sobelx = cv2.Sobel(img_gaussian,cv2.CV_8U,1,0,ksize=5)
img_sobely = cv2.Sobel(img_gaussian,cv2.CV_8U,0,1,ksize=5)
img_sobel = img_sobelx + img_sobely
```

```python
#prewitt: first deravative operation. calculate the image gradient.
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
#Horizontal mask and It is equivalent to taking the gradient into vertical direction.
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
#Vertical mask and It is equivalent to taking the gradient into horizontal direction.
img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)
#cv2.filter2D(src, ddepth, kernel)
#ddepth = -1; means the same depth of src image will be for the destination image.
#The source(img_gaussian image on which to apply the fitler(Kernal). It is a matrix that represents the image in pixel intensity values. ddepth: It is the desirable depth of destination image.
```

```python
# Very sensitive to noise, gives double edges but gives which sie point lies, also accurate location of edges using this operator. 
image_laplacian = cv2.Laplacian(img_gaussian, cv2.CV_8U, -1, ksize=5)
io.imshow(image_laplacian)
```
