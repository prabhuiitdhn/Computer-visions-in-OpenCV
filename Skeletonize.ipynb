```python
from skimage.morphology import skeletonize
import matplotlib.pyplot as plt
from skimage.util import invert
import skimage.io
```

```python
image_path ='C:/Users/uib43225/PycharmProjects/DSAlgo/computer vision'
image = image_path + '/cat_224x224.jpg'
```

```python
cat = skimage.io.imread(image)
skimage.io.imshow(cat)
```

```python
# invert an image. invert the intensity of input image, so that the dtype maximum is now dtype minimum and vice versa
image = invert(cat)
```

```python
skimage.io.imshow(image)
```

```python
skeleton = skeletonize(cat)
```

```python
skimage.io.imshow(skeleton)
```

```python
fig, axes = plt.subplots(nrows =1, 
                        ncols = 2,
                        figsize =(8, 4),
                        sharex = True, 
                        sharey = True)
```

```python
ax = axes.ravel()
```

```python
ax[0].imshow(cat, cmap=plt.cm.gray)
```

```python
ax[0].axis('off')
```

```python
ax[0].set_title('original', fontsize=20)
```

```python
ax[1].imshow(skeleton, cmap=plt.cm.gray)
```

```python
ax[1].axis('off')
ax[1].set_title('skeleton', fontsize=20)
```

```python
fig.tight_layout()
plt.show()
```
